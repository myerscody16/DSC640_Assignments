```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("world-population.xlsm")
data.head()
```

```python
plt.plot(data["Year"], data["Population"])
plt.title("World Population by Year")
plt.xlabel("Year")
plt.ylabel("Population in billions")
```

```python
plt.step(data["Year"], data["Population"])
plt.title("World Population by Year")
plt.xlabel("Year")
plt.ylabel("Population in billions")
```
