```python
x = np.arange(-5.0, 10.0, 1.25)
y = np.arange(-5.0, 10.0, 1.3)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
plt.contour(X, Y, Z)
```
