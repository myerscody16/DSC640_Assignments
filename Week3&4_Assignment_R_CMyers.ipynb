```r
library("ggplot2")
library("readxl")
```

```r
data <- read_excel("world-population.xlsm")
head(data)
```

```r
ggplot(data=data, aes(x=Year, y=Population)) + 
    geom_line() +
    labs(title="World Population by Year") + 
    theme(plot.title = element_text(hjust = 0.5))
```

```r
ggplot(data=data, aes(x=Year, y=Population)) + 
    geom_step() +
    labs(title="World Population by Year") + 
    theme(plot.title = element_text(hjust = 0.5))
```
